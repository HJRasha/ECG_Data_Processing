```python
import os
import pandas as pd

# Define the folder containing the CSV files
folder_path = r"D:\Varsity\Thesis\ECG dataset\af"

# List all CSV files in the folder
csv_files = [file for file in os.listdir(folder_path) if file.endswith('.csv')]

# Loop through each CSV file and change the header
for file_name in csv_files:
    file_path = os.path.join(folder_path, file_name)
    
    # Read the CSV file
    df = pd.read_csv(file_path)
    
    # Define the new header names as a list
    new_header = ['Time', 'ECG1']  # Replace with your desired column names
    
    # Rename the columns
    df.columns = new_header
    
    # Save the updated DataFrame back to the same CSV file
    df.to_csv(file_path, index=False)

    #print(f'Header updated for file: {file_name}')
```
